```python
import cv2
import numpy as np
import pyautogui
import imutils

import pyttsx3

    
def name():
    engine=pyttsx3.init()
    engine.say("my name is rahaf")
    engine.runAndWait()
    
def age():
    engine=pyttsx3.init()
    engine.say("10")
    engine.runAndWait()
    
    
cap=cv2.VideoCapture(0)

while True:
    _,frame=cap.read()
    frame=cv2.flip(frame,1)
    frame=imutils.resize(frame,height=700,width=900)
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    
    lowerred=np.array([131,90,106])
    upperred=np.array([255,255,255])
    
    lowblue=np.array([40,150,166])
    upperblue=np.array([255,255,255])
    
    red_mask=cv2.inRange(hsv,lowerred,upperred)
    blue_mask=cv2.inRange(hsv,lowblue,upperblue)
    
    cv2.rectangle(frame,(0,0),(200,150),(0,0,255),1)
    cv2.putText(frame,'name',(70,80),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3,cv2.LINE_AA)
    
    
    cv2.rectangle(frame,(300,0),(500,150),(0,0,255),1)
    cv2.putText(frame,'Age',(350,100),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),3,cv2.LINE_AA)
    
    
    
    
    
    
    
   
    contours , hierachy=cv2.findContours(red_mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    contours=sorted(contours,key=lambda x:cv2.contourArea(x),reverse=True)
    
    
    
    
    for cnt in contours:
        (x,y,w,h)=cv2.boundingRect(cnt)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        print(x,y)
        if x>0 and y>0 and x<200 and y<150:
            name()
        elif x>200 and y >0 and x<500 and y<150:
            age()
        break
    
   
    
    
    
    cv2.imshow("frame",frame)
    key=cv2.waitKey(1)
    if cv2.waitKey(1) & 0xFF== ord('q') :
        break

cap.release()
cv2.destroyAllWindows()
    

```
